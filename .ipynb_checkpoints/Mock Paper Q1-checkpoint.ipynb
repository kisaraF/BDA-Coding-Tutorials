## Mock Question 01

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('C:/Users/USER/Downloads/BDA/Dataset2_Mushrooms.csv')
df.head()
```

```python
df.isna().sum()
```

```python
df['poisonous'].value_counts()
```

```python
#Splittig features and labels
X = df.drop('poisonous', axis=1)
y = df['poisonous']
```

```python
#Train test splits
from sklearn.model_selection import train_test_split

X_train_i, X_test_i, y_train_i, y_test_i = train_test_split(X, y, test_size= 0.3)
```

```python
#Let's get the decision tree
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier()
model.fit(X_train_i, y_train_i)
```

```python
#Let's get the predicitions
y_preds = model.predict(X_test_i)
y_preds
```

```python
#Let's see the accuracy of the model
from sklearn.metrics import accuracy_score

accuracy_score(y_test_i, y_preds)
```

```python
#Let's get the confusion matrix
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm_i = confusion_matrix(y_test_i, y_preds)
cm_i
```

```python
#Visualizing it
display = ConfusionMatrixDisplay(confusion_matrix= cm_i, display_labels= model.classes_)
display.plot()
```

## Question 2

```python
data = pd.read_csv('C:/Users/USER/Downloads/BDA/Dataset_Phishing.csv')
data.head()
```

```python
#Finding if there's any missing values
data.isna().sum()
```

```python
#We can drop the id column
data_i = data.drop('id', axis=1)
len(data_i)
```

```python
data_i['CLASS_LABEL'].value_counts()
```

We can select `CLASS_LABEL` as the label column in our dataset

```python
import numpy as np
#Let's split features and the labels
X_ii = data_i.drop('CLASS_LABEL', axis=1)
y_ii = data_i['CLASS_LABEL']

#Let's check if it's done properly
X_ii.shape, y_ii.shape
```

```python
#Let's split the train test splits for the model training
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_ii, y_ii, test_size = 0.35)
```

```python
#Let's instantiate the model class object and train the model by fitting the data
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()

#Fitting the data
tree_model.fit(X_train, y_train)
```

```python
#Let's get the predictions
y_preds= tree_model.predict(X_test)
y_preds
```

```python
len(y_preds)
```

```python
#Checking the accuracy of the model
from sklearn.metrics import accuracy_score

round(accuracy_score(y_test, y_preds),2)
```

Let's build a `confusion matrix` to see accuracy in detail

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_preds)
cm
```

```python
cm.shape
```

```python
#Visualizing the confusion matrix
display = ConfusionMatrixDisplay(confusion_matrix= cm, display_labels= tree_model.classes_)
display.plot()
```
